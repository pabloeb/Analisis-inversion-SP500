5- DATAFRAME CON KPI´S FUNDAMENTALES DEL SECTOR ENERGY

```python
# Se modifica el .csv importado de web page 

df_Fundamentals = pd.read_csv(r"Archivos .ods/Fundamentals_Energy_www.csv", sep = ";")
df_Fundamentals.head(2)
```

```python
df_Fundamentals.info()
```

```python
cols_to_drop = df_Fundamentals.columns[df_Fundamentals.columns.str.startswith('Unn')]

# Eliminar las columnas seleccionadas
df_Fundamentals.drop(cols_to_drop, axis=1, inplace=True)
# Se elimina el símbolo porcentaje y se pasa columna a float
df_Fundamentals['DIVIDEND YIELD'] = df_Fundamentals['DIVIDEND YIELD'].str.replace("%", "").astype(float)
df_Fundamentals.info()

```

```python
# Se cambian los nombres de las columnas

df_Fundamentals = df_Fundamentals.rename(columns={
    'MARKET CAPITALIZATION $M': 'MARKET CAPITALIZATION',
    'PRICE/EARNINGS TTM': 'PRICE/EARNINGS',
    'EARNINGS PER SHARE TTM': 'EARNINGS PER SHARE'
})
```

Se define un ranking de empresas del sector ENERGY tomando tres KPIs denominados "Fundamentals", con diferentes ponderaciones para calcular un nuevo KPI.El KPI DIVIDEND YIEL o se pondero 10% por sobre los otros dos KPIs en función de que en la situación actual de mercado, donde la rentabilidad de las acciones son magras respecto a otras inversiones,  

```python
# Definir las columnas que se utilizarán para asignar puntos a las empresas
columns = ['MARKET CAPITALIZATION' , 'PRICE/EARNINGS', 'DIVIDEND YIELD']

# Recorrer cada columna
for col in columns:
    # Obtener la posición de cada empresa dentro de la columna
    rank = df_Fundamentals[col].rank(method='min')
    # Calcular el máximo valor de la columna y restarlo a todas las posiciones
    rank = rank.max() - rank + 1
    # Asignar el ranking a una nueva columna
    df_Fundamentals[col+'_RANK'] = rank.astype(int)

# Sumar los rangos de las tres columnas
df_Fundamentals['TOTAL_RANK'] =1.05* df_Fundamentals[columns[0]+'_RANK'] + df_Fundamentals[columns[1]+'_RANK'] + df_Fundamentals[columns[2]+'_RANK']

# Ordenar el dataframe por la puntuación total de menor a mayor
df_Fundamentals = df_Fundamentals.sort_values(by='TOTAL_RANK', ascending=True)

# Resetear el índice del dataframe
df_Fundamentals = df_Fundamentals.reset_index(drop=True)

#Se fuerza al indice a empezar en 1, para utilizar la columna indice como indicativo de posición de ranking
df_Fundamentals.index = df_Fundamentals.index + 1

df_Fundamentals["Ranking"] = df_Fundamentals.index

df_Fundamentals['DIVIDEND YIELD'].dtype
```
